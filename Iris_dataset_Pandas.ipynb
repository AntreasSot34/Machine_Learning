```python
# Use the Data Frame from Panda Library to extract data from csv and excel files
import pandas as pd
# Creating a Variety of visualizations: Line, Bar, Histograms, Scatter, Pie, SubPlots, Customizations
import matplotlib.pyplot as plt
# Provides utilities to load and generate various datasets(Benchmarking,EDA, Model Testing)
from sklearn.datasets import load_iris
"""
Overview of the Iris Dataset
The Iris dataset consists of 150 samples from three species of Iris flowers: Iris setosa, Iris versicolor, and Iris virginica. Each species has 50 samples. The dataset contains the following features (or attributes) for each sample:

Sepal Length (in centimeters)
Sepal Width (in centimeters)
Petal Length (in centimeters)
Petal Width (in centimeters)
These four features are used to describe the physical characteristics of the flowers.

The Target Variable
Species: This is the target variable that indicates the species of each flower sample. It has three possible values corresponding to the three Iris species:
Iris setosa
Iris versicolor
Iris virginica
"""
```

```python
# Load the Iris dataset from sklearn
iris = load_iris()
iris
```

```python
# Create a Pandas DataFrame from the data
data = iris.data
# Feature are the headers of the table
feature_names = iris.feature_names
target = iris.target
#Connect the targets with the data
target_names = iris.target_names

df = pd.DataFrame(data, columns=feature_names)
df['species'] = [target_names[i] for i in target]
```

```python
# Get information about the data (columns, data types, non-null values)
print("\nData Information:")
print(df.info())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(df.describe().to_markdown(numalign="left", stralign="left"))
```

```python
# Sorting
print("\nSorted by Petal Length (descending):")
print(df.sort_values(by='petal length (cm)', ascending=False).head().to_markdown(index=False, numalign="center", stralign="center"))
```

```python
# Simple Visualization (Corrected): Mathplot Lib
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c= iris.target)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs. Sepal Width by Species') # Added "by Species" to title for clarity
plt.show()
```
